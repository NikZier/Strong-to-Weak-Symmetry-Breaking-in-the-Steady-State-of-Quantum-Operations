```python
colors2 = plt.get_cmap("Greens")(np.linspace(0.1, 1.0, 5))
colors = plt.get_cmap("Blues")(np.linspace(0.1, 1.0, 5))
fig, ax = plt.subplots(1)

s=0.5
ps = s*data_L64[2]
ax.plot(ps,R2_L8_sig,color=colors[1],zorder=1, linewidth = 2.5)
L8s=ax.scatter(ps,R2_L8_sig, marker="o", color=colors[1], edgecolor=colors[0], linewidth = 1, label = "L=8", s = 90)

ax.plot(ps,R2_L8_Q,color=colors2[1],zorder=1, linewidth = 2.5)
L8Q=ax.scatter(ps,R2_L8_Q, marker="s", color=colors2[1], edgecolor=colors2[0], linewidth = 1, label = "L=8", s = 90)

ax.plot(ps,R2_L16_sig,color=colors[2],zorder=1, linewidth = 2.5)
L16s=ax.scatter(ps,R2_L16_sig, marker="o", color=colors[2], edgecolor=colors[0], linewidth = 1, label = "L=16", s = 90)

ax.plot(ps,R2_L16_Q,color=colors2[2],zorder=1, linewidth = 2.5)
L16Q=ax.scatter(ps,R2_L16_Q, marker="s", color=colors2[2], edgecolor=colors2[0], linewidth = 1, label = "L=16", s = 90)

ax.plot(ps,R2_L32_sig,color=colors[3],zorder=1, linewidth = 2.5)
L32s=ax.scatter(ps,R2_L32_sig, marker="o", color=colors[3], edgecolor=colors[0], linewidth = 1, label = "L=32", s = 90)


ax.plot(ps,R2_L32_Q,color=colors2[3],zorder=1, linewidth = 2.5)
L32Q=ax.scatter(ps,R2_L32_Q, marker="s", color=colors2[3], edgecolor=colors2[0], linewidth = 1, label = "L=32", s = 90)

ax.plot(ps[3:],R2_L64_sig[3:],color=colors[4],zorder=1, linewidth = 2.5)
L64s=ax.scatter(ps[3:],R2_L64_sig[3:], marker="o", color=colors[4], edgecolor=colors[0], linewidth = 1, label = "L=64", s = 90)

ax.plot(ps[3:],R2_L64_Q[3:],color=colors2[4],zorder=1, linewidth = 2.5)
L64Q=ax.scatter(ps[3:],R2_L64_Q[3:], marker="s", color=colors2[4], edgecolor=colors2[0], linewidth = 1, label = "L=64", s = 90)

legend1 = ax.legend(handles=[L8s,L16s,L32s,L64s], loc="upper right", title=r"$\sigma$-$\sigma$ correlations", frameon = False, fontsize = 12)
ax.add_artist(legend1)

legend1 = ax.legend(handles=[L8Q,L16Q,L32Q,L64Q], loc="center right", title=r"$Q$-$Q$ correlations", frameon = False, fontsize = 12)
ax.add_artist(legend1)


ax.set_xlabel(r"$ps$", fontsize = 18)
ax.set_yticks([0,0.25,0.5])
ax.set_xticks([0,2,4,6])
ax.set_yticklabels(["0","0.25","0.5"])
ax.tick_params(labelsize = 16)
plt.gcf().subplots_adjust(left=0.20)
plt.gcf().subplots_adjust(bottom=0.15)
ax.set_ylabel(r"$C^{sw}(\frac{N}{4},\frac{3N}{4})$", fontsize = 18)
fig.savefig(r"C:\Users\nzier\Documents\QuantumActiveMatter\QuantumCircuits\figs\Renyi2_MPS_S3_legend.pdf", format = "pdf")
```
